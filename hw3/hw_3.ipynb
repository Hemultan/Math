### Задание 1

Дан объект в $2D$ пространстве

````python
import numpy as np
import matplotlib.pyplot as plt
import math

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````python
###Уменьшить объект в два раза
B = np.array([[0.5, 0, 0], [0, 0.5, 0]])
mult = B@A.T
res = mult.T
x = res[:,0]
y = res[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
### Повернуть на 130 градусов (здесь примерно близко к 130)
rotate = np.array([[np.cos(7*math.pi/6), -np.sin(7*math.pi/6)], [np.sin(130), np.cos(7*math.pi/6)]])
mult_1 = rotate@res.T
rot = mult_1.T
x = rot[:,0]
y = rot[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
###Отразить объект относительно прямой y=x
line = np.array([[-1000, -1000], [1000, 1000]])
x1 = line[:,0]
y1 = line[:,1]

plt.plot(x1, y1)
plt.plot(-x, -y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

### Задание 2

Найдите спектральное разложение матрицы:
```
m = np.array([[1, 2],
              [2, 3]])
print(m)
```



````python
m = np.array([[1, 2], [2, 3]])
````

````python
values, vectors = np.linalg.eig(m)
````

````python
### Матрица, столбцы которой являются собственными векторами
S = vectors
S
````

````python
### Транспонированная матрица, столбцы которой являются собственными векторами
S_1 = S.T
S_1
````

````python
### Диагональная матрица, с собственными значенями
D = np.diag(values)
D
````

````python
### Проверка
np.dot(np.dot(S, D), S_1)
````
