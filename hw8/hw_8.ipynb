### Задание 1

В магазин привезли устройства с 3-х разных предприятий. 

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта. 

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```python
PA1 = 0.2
PA2 = 0.3
PA3 = 0.5
PBA1 = 0.1
PBA2 = 0.05
PBA3 = 0.2

PB = (PA1 * PBA1) + (PA2 * PBA2) + (PA3 * PBA3)

print(f"Вероятность того, что случайно купленная новая продукция окажется высшего сорта: {PB*100}%")
```

### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```python
###являются попарно независимыми
m1, m2, m3 = 0.5, 0.5, 0.5
PA1 = m1*m2
PA2 = m2*m3
PA3 = m1*m3

PA12 = PA1 * PA2
PA23 = PA2 * PA3
PA13 = PA1 * PA3

PA123 = PA1*PA2*PA3

print(f"Попарная независомсть: {PA12, PA23, PA13}")
print(f"Совокупная независимость: {PA123}")
```

### Задание 3


Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения. 

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```python
### Нормальное распределение
"""
Параметры np.linspace(norm.ppf(0.33), norm.ppf(0.9), 100) - Первые два параметры это границы, из которых 
вычисляется некая дельта для дальнейшх преобразований. В целом похоже на область значений, доверительные интервалы 
и/или сигмы, так как речь идет о нормальном распределении.
За ними идет параметр, в нашем случае 100 - это количество используемых точек/значений в распределении.
"""

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)

x1 = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
plt.plot(x1, norm.pdf(x1), lw=3, label='x1')
plt.legend()
plt.show()

x2 = np.linspace(norm.ppf(0.33), norm.ppf(0.9), 100)
plt.plot(x2,norm.pdf(x2), lw=3, label='x2')
plt.legend()
plt.show()

x3 = np.linspace(norm.ppf(0.33), norm.ppf(0.2), 100)
plt.plot(x3, norm.pdf(x3), lw=3, label='x3')
plt.legend()
plt.show()

x4 = np.linspace(norm.ppf(0.3), norm.ppf(0.9), 10)
plt.plot(x4, norm.pdf(x4), lw=3, label='x4')
plt.legend()
plt.show()

x5 = np.linspace(norm.ppf(0.2), norm.ppf(0.5), 4)
plt.plot(x5, norm.pdf(x5), lw=3, label='x5')
plt.legend()
plt.show()
```

```python
### Экспоненциальное распределение
"""
np.linspace(expon.ppf(0.01), expon.ppf(0.99), 100) - Первые два параметры это границы, из которых 
вычисляется некая дельта для дальнейшх преобразований. В целом похоже на область значений. 
Иногда изменяется рост/падение/монотонность
За ними идет параметр, в нашем случае 100 - это количество используемых точек/значений в распределении.

"""

from scipy.stats import expon
fig, ax = plt.subplots(1, 1)

e1 = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 100)
plt.plot(e1, norm.pdf(e1), lw=3, color='m', label='e1')
plt.legend()
plt.show()

e2 = np.linspace(expon.ppf(0.4), expon.ppf(0.99), 100)
plt.plot(e2, norm.pdf(e2), lw=3, color='m', label='e2')
plt.legend()
plt.show()

e3 = np.linspace(expon.ppf(0.05), expon.ppf(0.2), 100)
plt.plot(e3, norm.pdf(e3), lw=3, color='m', label='e3')
plt.legend()
plt.show()

e4 = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 10)
plt.plot(e4, norm.pdf(e4), lw=3, color='m', label='e4')
plt.legend()
plt.show()

e5 = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 4)
plt.plot(e5, norm.pdf(e5), lw=3, color='m', label='e5')
plt.legend()
plt.show()
```

```python
### Экспоненциальное распределение
"""
np.linspace(t.ppf(0.01, df1), t.ppf(0.99, df1), 100) - Первые два параметры это границы, из которых 
вычисляется некая дельта для дальнейшх преобразований. В целом похоже на область значений. Изменение параметра df
растягивает купол по оси x, увеличивая/уменьшая доверительные интервалы и/или сигмы
За ними идет параметр, в нашем случае 100 - это количество используемых точек/значений в распределении.

"""

from scipy.stats import t
fig, ax = plt.subplots(1, 1)

df1 = 2
df2 = 6

t1 = np.linspace(t.ppf(0.01, df1), t.ppf(0.99, df1), 100)
plt.plot(t1, norm.pdf(t1), lw=3, color='y', label='t1')
plt.legend()
plt.show()

t2 = np.linspace(t.ppf(0.01, df2), t.ppf(0.99, df2), 100)
plt.plot(t2, norm.pdf(t2), lw=3, color='y', label='t2')
plt.legend()
plt.show()

t3 = np.linspace(t.ppf(0.4, df1), t.ppf(0.2, df1), 100)
plt.plot(t3, norm.pdf(t3), lw=3, color='y', label='t3')
plt.legend()
plt.show()

t4 = np.linspace(t.ppf(0.01, df2), t.ppf(0.99, df2), 10)
plt.plot(t4, norm.pdf(t4), lw=3, color='y', label='t4')
plt.legend()
plt.show()

t5 = np.linspace(t.ppf(0.01, df1), t.ppf(0.99, df1), 4)
plt.plot(t5, norm.pdf(t5), lw=3, color='y', label='t5')
plt.legend()
plt.show()
```
